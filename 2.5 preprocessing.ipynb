# Cleaning answers

```python
# Remove newlines to avoid issues
df['answer_clean'] = df['answer'].str.replace('\n', '', regex=False)
```

```python
# Find rows that have other (non-newline) issues
from json import JSONDecodeError

def try_parse(x):
    try:
        return json.loads(x)
    except JSONDecodeError:
        return None

df['answer_test'] = df['answer_clean'].apply(try_parse)

df['parse_failed'] = df['answer_clean'].apply(lambda x: pd.isna(try_parse(x)))

# Count how many are broken
broken = df[df['parse_failed']]
broken # 169 rows where parsing failed (answer entry wasn't written to the end, breaks off in the middle)

# Remove broken rows
df = df.drop(broken.index)
df.shape
```

```python
# Unnest columns
df['answer_parsed'] = df['answer_clean'].apply(json.loads)
df_answers = pd.json_normalize(df['answer_parsed'])
df_answers
```

```python
df = pd.concat([
    df[['title', 'author', 'fmc_name']].reset_index(drop=True),
    df_answers.reset_index(drop=True)
], axis=1)
df
```

```python
df.to_csv('data/df_preprocessed.csv')
```
