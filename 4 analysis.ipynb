2. Nur positiven stereotypicality sum score angucken: schreiben Mnner mehr stereotypisch?
3. Nur negativen, anti-stereotypicality sum score angucken: schreiben Mnner weniger anti-stereotypisch?

```python
# Recode stereotpyical (even) columns -----------------------------------

# 1. Identify even-numbered *.answer columns
stereotypical_cols = [col for col in df.columns if col.endswith('.answer') and int(col[1:col.find('.')]) % 2 == 0]

# 2. Recode only 99 to NaN
df[stereotypical_cols] = df[stereotypical_cols].replace({99: np.nan})

# Check:
print(df[stereotypical_cols].head())
```

```python
# Recode  anti-stereotpyical (odd) columns -----------------------------------

# 1. Identify odd-numbered *.answer columns
anti_stereotypical_cols = [col for col in df.columns if col.endswith('.answer') and int(col[1:col.find('.')]) % 2 == 1]

# 2. Recode 1 to -1 and 99 to NaN
df[anti_stereotypical_cols] = df[anti_stereotypical_cols].replace({1: -1, 99: np.nan})

# Check:
print(df[anti_stereotypical_cols].head())
```

```python
# Identify ALL answer columns
answer_cols = [col for col in df.columns if col.endswith('.answer')]
```

```python
# Create sum scores 

# --- TOTAL SUM ---
df['total_sum'] = df[answer_cols].sum(axis=1, skipna=True)

# --- POSITIVE SUM (only 1s from even-numbered answer columns) ---
df['pos_sum'] = df[stereotypical_cols].apply(lambda row: (row == 1).sum(), axis=1)

# --- NEGATIVE SUM (only -1s from odd-numbered answer columns) ---
df['neg_sum'] = df[anti_stereotypical_cols].apply(lambda row: (row == -1).sum(), axis=1)
```

```python
# Calculate % of NAs in even-numbered .answer columns
even_na_percent = df[stereotypical_cols].isna().mean().mean() * 100

# Calculate % of NAs in odd-numbered .answer columns
odd_na_percent = df[anti_stereotypical_cols].isna().mean().mean() * 100

print(f" Average % of NAs in even-numbered questions: {even_na_percent:.2f}%")
print(f" Average % of NAs in odd-numbered questions: {odd_na_percent:.2f}%")
```

```python
# NA percentage per even-numbered column
even_na_percents = df[stereotypical_cols].isna().mean() * 100
print(" NA % per even-numbered question:")
print(even_na_percents.round(2))

# NA percentage per odd-numbered column
odd_na_percents = df[anti_stereotypical_cols].isna().mean() * 100
print("\n NA % per odd-numbered question:")
print(odd_na_percents.round(2))
```
