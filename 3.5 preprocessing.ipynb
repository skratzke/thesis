```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/df_final.csv")
df
```

```python
# Remove authors where gender is unclear/non-binary (==99)
df = df[df['author_gender'] != 99].reset_index(drop=True).copy()
```

```python
# remove years that have a 0 or 99 
df = df[df['year'] > 99].reset_index(drop=True).copy()
df
```

```python
# Recode stereotpyical (even) columns -----------------------------------

# 1. Identify even-numbered *.answer columns
stereotypical_cols = [col for col in df.columns if col.endswith('.answer') and int(col[1:col.find('.')]) % 2 == 0]

# 2. Recode  99 to NaN
df[stereotypical_cols] = df[stereotypical_cols].replace({99: np.nan})

# Check:
print(df[stereotypical_cols].head())
```

```python
# Recode  anti-stereotpyical (odd) columns -----------------------------------

# 1. Identify odd-numbered *.answer columns
anti_stereotypical_cols = [col for col in df.columns if col.endswith('.answer') and int(col[1:col.find('.')]) % 2 == 1]

# 2. Recode  99 to NaN
df[anti_stereotypical_cols] = df[anti_stereotypical_cols].replace({99: np.nan})

# Check:
print(df[anti_stereotypical_cols].head())
```

```python
# Identify ALL answer columns
answer_cols = [col for col in df.columns if col.endswith('.answer')]
```

```python
# Anzahl beurteilte (non‑NA) Items pro Zeile
n_pos = df[stereotypical_cols].notna().sum(axis=1)
n_neg = df[anti_stereotypical_cols].notna().sum(axis=1)

# Anzahl Hits (1er) pro Zeile
sum_pos = df[stereotypical_cols].eq(1).sum(axis=1)
sum_neg = df[anti_stereotypical_cols].eq(1).sum(axis=1)
```

```python
print(n_pos, sum_pos, sum_neg)
```

```python
# Proportion der Hits nur unter den Beurteilten
df['prop_pos'] = sum_pos / n_pos
df['prop_neg'] = sum_neg / n_neg

# Differenz der Proportionen
df['diff_prop'] = df['prop_pos'] - df['prop_neg']
```

Now we ONLY take into consideration items that are not NA. That means, if we have a row where all but 1 item is NA, and that one item is 1, then prop_pos would be 100%. 
This really assumes that there is a clear difference between a NA and a 0, which is something that should be discussed.
