````python
# Lowercase everything just to be safe
df['main_genre_clean'] = df['main_genre'].str.lower().str.strip()

# Map genres to master categories
df['master_genre'] = df['main_genre_clean'].map(genre_mapping)

# Optional: Label unmatched genres as 'Other'
df['master_genre'] = df['master_genre'].fillna('Other')

# Preview the result
df[['main_genre', 'master_genre']]
````

````python
# Look at 'Fiction' books and see what their other genres are 
fiction_books = df[df['master_genre'] == 'Fiction'].copy()
fiction_books['genres']
````

````python
# Step 1: Get the second genre if it exists
fiction_books.loc[:, 'second_genre'] = fiction_books['genres'].apply(lambda x: x[1] if len(x) > 1 else None)

# Step 2: Map to master genre
fiction_books.loc[:, 'second_master_genre'] = fiction_books['second_genre'].map(genre_mapping)

# Step 3: Label unmatched genres as 'Other'
fiction_books.loc[:, 'second_master_genre'] = fiction_books['second_master_genre'].fillna('Other')
fiction_books[['genres', 'second_master_genre']]
````

````python
fiction_books['second_master_genre'].value_counts()
````

````python
df.loc[fiction_books.index, 'master_genre'] = fiction_books['second_master_genre']
````
