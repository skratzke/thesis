Goal: reduce all genre names into around 10 genre clusters

```python
import pandas as pd
```

```python
df = pd.read_csv('data/df_preprocessed.csv')
df
```

```python
df['main_genre'] = df['genres'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
```

```python
df['main_genre'].value_counts().reset_index(name='count').rename(columns={'index': 'main_genre'})
```

```python
df['two_genres'] = df['genres'].apply(lambda x: x[:2] if isinstance(x, list) and len(x) > 0 else None)
df['two_genres']
```

```python
df['two_genres'].value_counts().reset_index(name='count').rename(columns={'index': 'two_genres'})
```

###########################################

```python
import ast
from sklearn.preprocessing import MultiLabelBinarizer

# Step 1: Convert stringified lists to actual lists (if needed)
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Step 2: Only keep the first two elements of each genre list
df['genres_trimmed'] = df['genres'].apply(lambda x: x[:2] if isinstance(x, list) else [])

# Step 3: One-hot encode the trimmed genres
mlb = MultiLabelBinarizer()
genre_dummies = mlb.fit_transform(df['genres_trimmed'])

# Step 4: Create a DataFrame with genre dummy columns
genre_df = pd.DataFrame(genre_dummies, columns=mlb.classes_, index=df.index)
genre_df
```

```python
import re

# Clean column names by replacing non-alphanumeric characters with underscores
genre_df.columns = [
    re.sub(r'\W+', '_', col).strip('_').lower()  # optional: also convert to lowercase
    for col in genre_df.columns
]
genre_df
```

```python
genre_df.to_csv('data/genre_one_hot.csv')
```
